## Introduction
The Helium Neon (He-Ne) laser relies upon collisions between Ne and metastable Helium atoms, which are contained as plasma in a narrow bore tube. These collisions cause Ne electrons to be raised up to the 5s or 4s level. Excited atoms then decay to a lower energy level, emitting a photon. This photon, when coming into contact with another excited atom, can induce an identical transition. In this way, simulated emission of radiation is achieved.

*FIG. 1. Energy level diagram for He-Ne Laser. Diagram sourced from "Optics Guide 4", Melles Griot (1988)[1].*

Mirrors on both sides of the tube reflect photons through the bore, thus increasing simulated emission creating light amplification. The mirrors form an optical cavity that reflects only the desired wavelength. The other wavelengths escape through the cavity. Around $1%$ of the desired wavelength is allowed to escape through one of the mirrors, this being the usable beam.

## Method

The method for this experiment was drawn largely from the student notes. In order to fully investigate operation and properties of the He-Ne Laser, the experiment was broken down into four parts:  
  
  i. Experimental derivation of He-Ne Laser wavelength.   
  ii. Spectroscopy     
  iii. Investigation of the spatial filter.  
  iv. Determination of beam divergence.  

Prior to and measurements, the He-Ne Laser turned on and levelled. 

### i. Experimental Derivation of He-Ne Laser Wavelength. 
In accordance with the student notes[2], the diffraction grating was mounted onto the optical rail $a = (350.0 \pm 0.5) mm$ away from the screen. Special care was taken to align the laser to the grating before measuring the separation between the zeroth and first order maxima could be easily measured with a metre ruler. 

### ii. Spectroscopy.
Without turning off the laser, the "SpectraSuite" software was opened and set up according to the student notes[3]. The spectrometer probe was then used to measure the spectrum of 6 different light sources: laser output, laser cavity, helium plasma, neon plasma, incandescent light bulb and incandescent light bulb through laser tube. For each light source, the spectra was recorded with "SpectraSuite" and exported as a .txt file. In order to prevent saturation, the probe was kept a suitable distance from the light source at all times. 

### iii. Investigation of the Spatial Filter.
In keeping with the student notes[4], the Spiricon beam profiler was connected to the computer via usb and mounted on the optical rail approximately $30cm$ from the laser aperture. After running the relevant software, "PLL 4 - Spatial Filter", the laser beam was adjusted so that it was centred in the image box. The exposure level was set and maintained so that the maximum pixel intensity remained below $100$. The spatial filter assembly was then mounted onto the optical rail and adjusted so that beam passed through its centre. Using "LabVIEW" an intensity distribution was measured and exported firstly for no pinhole, then $100$, $50$ and $25\mu m$ spatial filters. 

### iv. Determination of Beam Divergence.  
After removing the spatial filter assembly from the optical rail, the Spirison beam profiler was moved $(10.0 \pm 0.5)cm$ away from the laser aperture. The "PLL 3 - Beam Divergence" program was run and the FWHM recorded. This measurement was then repeated for a series of distances from $(10.0 - 30.0 \pm 0.5)cm$ at intervals of $(2.0 \pm 0.5)cm$, then $(5.0 \pm 0.5)cm$ intervals between $(30.0 - 130.0 \pm 0.5)cm$. A FWHM vs distance from the laser aperture plot was created and exported.  



## Results and Analysis

### Deriving the He-Ne Laser Wavelength  
  
The experimental value of the wavelength of red light can be calculated according the equation $d\sin\theta = n\lambda$ where $d$ is the grating spacing $d = 1.69Î¼m$ (given), $\theta$ is the angle of diffraction and $n$ is the order of grating $n = 1$. A right angle is formed between the grating, screen and diffraction maxima. As such, the Pythagorean theorem $c = \sqrt{a^2 + b^2}$, where c is the hypotenuse, a is the distance between the grating and the screen $(350.0 \pm 0.5)mm$ and b is the distance between the zeroth and first order maxima $(147.0 \pm 0.5)mm$, can be used to find the value of $\sin\theta$. By re-arranging, the experimental value of the wavelength of red light $\lambda = (654 \pm 3)nm$. A more detailed derivation of this result (including uncertainty propagation) can be found in the lab book.
     
The accepted value of the wavelength for red light is given to be $\lambda = 633nm$. This leaves a discrepancy and relative error of $(\pm 9.5)nm$ and $0.33%$ respectively, which is out of the bounds of our calculated uncertainty $(\pm 3)nm$. A high relative error and low relative uncertainty $0.0046%$ suggests a precise but inaccurate experiment and implies the existence of systematic errors. One possible source of error may have been from a misaligned laser. Another possibility is that we underestimated the uncertainty of a metre ruler $(\pm 0.5)mm$ and our ability to read a measurement within $1mm$ in low light. 

### Spectroscopy
  
In order to effectively investigate emitted spectra and the effect laser cavity mirrors, the spectra of the six light sources were grouped into three pairs for comparison. The resultant plots and analysis can be seen below. 

```python
# Plotting laser vs laser cavity

import matplotlib.pyplot as plt 
import numpy as np

x, y = np.loadtxt('tiny_hole.csv', delimiter=',', unpack=True)
b, c = np.loadtxt('laser.csv', delimiter=',', unpack=True)
plt.plot(x,y, color='b', label='Laser Cavity')
plt.plot(b,c, color='r', label='Laser')


plt.title('Laser vs Laser Cavity')  
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity') 
plt.legend()
plt.show()
```

*FIG. 2. Plot comparing spectra of the laser cavity and laser output.*

The wavelength of the laser output, $lambda = 633nm$, can be extracted directly from the red peak in FIG. 2.. It should be noted that a intense peak of the same wavelength does not exist in the laser cavity spectra. Instead, intense blue peaks can be seen at $\lambda = (502, 588, 641, 668, 707)nm$. These wavelengths relate to spectra of the He and Ne spectra, see FIG. 2.. This observed phenomena agrees with the workings of a laser as described in the introduction. The $633nm$ peak seen in the laser output spectra represents the $1%$ of light that is allowed to escape from one of the mirrors in the bore tube. Whilst the many blue peaks relate to unwanted wavelengths allowed to escape as cavity losses.

```python
# Incandescent Light Bulb vs. Incandescent Light Bulb through Laser Tube

import matplotlib.pyplot as plt 
import numpy as np

x, y = np.loadtxt('desk_lamp.csv', delimiter=',', unpack=True)
b, c = np.loadtxt('laser_tube_large.csv', delimiter=',', unpack=True)
d, e = np.loadtxt('laser_tube_medium.csv', delimiter=',', unpack=True)
f, g = np.loadtxt('laser_tube_small.csv', delimiter=',', unpack=True)
plt.plot(x,y, color='r', label='Incandescent Bulb')
plt.plot(b,c, color='b', label='Large Laser Tube')
plt.plot(d,e, color='g', label='Medium Laser Tube')
plt.plot(f,g, color='m', label='Small Laser Tube')


plt.title('Incandescent Light Bulb vs. Incandescent Light Bulb Through Laser Tube')  
plt.xlabel('Wavelength')
plt.ylabel('Intensity') 
plt.legend()
plt.show()
```

*FIG. 4. Plot comparing spectra of an incandescent light bulb and an incandescent light bulb through three sizes of laser tubes: small, medium and large.*   
  
FIG. 4. demonstrates the effect of a laser tube on an incandescent light bulb. By itself, an incandescent bulb emits a continuous spectrum of red, orange, yellow, green, blue, indigo and violet (ROYGBIV) light. In FIG. 4., this is represented by the red line. However, when filtered through any of the three laser tubes, wavelengths between $580nm$ and $720nm$ are not present. Similarly to the He-Ne laser, this phenomena can be explained by the effect of the reflectivity of the laser mirrors. These mirrors trap wavelength's between $580nm$ and $720nm$, whilst allowing wavelengths outside of that range to escape.
